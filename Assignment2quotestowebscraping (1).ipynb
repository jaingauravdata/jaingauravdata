```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
# list to store scraped data
all_quotes = []
```

```python
# this part of the url is constant
base_url = "http://quotes.toscrape.com/page/1"
```

```python
for i in range(1,6):
    url=f"http://quotes.toscrape.com/page/{i}"
    response=requests.get(url)
    response=response.content
    soup=BeautifulSoup(response,'html.parser')
    quotes=soup.find_all(class_='quote')
    for quote in quotes:
        all_quotes.append({
            "text": quote.find(class_="text").get_text(),
            "author": quote.find(class_="author").get_text(),
            "bio-link": quote.find("a")["href"]
        })
all_quotes
```

```python
# We will use dataframe to show all three values in a single dataframe.
df=pd.DataFrame(all_quotes,columns=['text','author','bio-link'])
df
```

```python
# Now  we will store this information into csv file.
df.to_csv('all_quotes.csv')
```

```python
# Now  we will store this information into excel file.
df.to_excel('all_quotes.xlsx')
```
